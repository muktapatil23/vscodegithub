### Agglomerative Clustering


```python
from warnings import filterwarnings
filterwarnings('ignore')
```

## Read the dataset

```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.columns = ['CustomerID', 'Gender', 'Age', 'Income', 'Spending']
```

```python
df.info()
```

```python
df.isna().sum()
```

### Create the clusters based on Age, Income and Spend (Seperate X results)


```python
X = df[['Age', 'Income', 'Spending']]
X.head()
```

### Perform preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre[0:5]
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=cols)
X_pre.head()
```

## Calculate silhoutte scores for different number of clusters

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
n = list(range(2, 16))
scores = []
for i in n:
    agc = AgglomerativeClustering(n_clusters=i)
    agc.fit(X_pre)
    lb = agc.labels_
    s = silhouette_score(X_pre, lb)
    scores.append(s)
```

```python
scores
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(n, scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouttee Score')
plt.title('Silhoutee Score for different number of clusters')
plt.grid()
plt.show()
```

## Create a final algorithm with 6 number of clusters

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
lb = model.labels_
lb
```

## Save the labels in X dataframe

```python
X['Label']=lb
```

```python
X
```

### View the differnet labels

```python
for i in range(6):
    print(f'Cluster Number : {i}\n')
    display(X[X['Label']==i])
    print('\n==========================================\n')
```

#### Plotting the dendogram

```python
from scipy.spatial import distance_matrix
dm = distance_matrix(X_pre, X_pre)
dm
```

```python
from scipy.cluster.hierarchy import linkage
linkages = linkage(dm)
```

```python
from scipy.cluster.hierarchy import dendrogram
Q = dendrogram(linkages)
```
